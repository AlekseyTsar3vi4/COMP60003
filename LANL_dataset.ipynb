```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the LANL dataset
lanl_df = pd.read_csv('lanl-auth-dataset-1-00.csv')

# Rename columns for consistency
lanl_df.columns = ['Timestamp', 'UserID', 'ComputerID']
# Extract necessary columns
lanl_df = lanl_df[['Timestamp', 'UserID', 'ComputerID']]

# Ensure no missing values
lanl_df.dropna(inplace=True)

# 1. Username (UserID) Distribution
plt.figure(figsize=(10, 6))
lanl_df['UserID'].value_counts().head(20).plot(kind='bar', color='skyblue')
plt.title('Top 20 Usernames - LANL Dataset')
plt.xlabel('Username')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# 2. Computer ID Distribution
plt.figure(figsize=(10, 6))
lanl_df['ComputerID'].value_counts().head(50).plot(kind='bar', color='lightgreen')
plt.title('Top 50 Computer IDs - LANL Dataset')
plt.xlabel('Computer ID')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Convert timestamps to hours
lanl_df['Hour'] = (lanl_df['Timestamp'] % (24 * 3600)) // 3600

# Create a histogram to distribute data evenly across 24 hours
plt.figure(figsize=(10, 6))
lanl_df['Hour'].value_counts(normalize=True).sort_index().plot(kind='bar', color='orange', alpha=0.7, edgecolor='black')
plt.title('Connection Time Distribution - LANL Dataset')
plt.xlabel('Hour of the Day')
plt.ylabel('Proportion of Connections')
plt.xticks(range(0, 24), labels=[f'{hour}:00' for hour in range(24)], rotation=45)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
